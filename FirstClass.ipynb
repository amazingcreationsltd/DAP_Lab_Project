1. Compute sum of digits of an integer (no string conversion).

```python
n = 134
sum = 0

while (n!=0):
  rem = int(n%10)
  # sum = sum*10 + rem
  sum = sum + rem
  n = int(n/10)

print(sum)
```

2. Compute product of digits (no string).

```python
n = 134
product = 1

while (n!=0):
  rem = int(n%10)
  # sum = sum*10 + rem
  product = product * rem
  n = int(n/10)

print(product)
```

3. Reverse a number (no string).

```python
n = 134
rev = 0

while (n!=0):
  rem = int(n%10)
  rev = rev*10 + rem
  # rev = rev + rem
  n = int(n/10)

print(rev)
```

4. Check palindrome number (no string).

```python
m = 121
temp = m
rev2 = 0

while (m!=0):
  rem = int(m%10)
  rev2 = rev2*10 + rem
  m = int(m/10)

if (rev2 == temp):
  print("Palindrome")
else:
  print("Not Palindrome")
```

5. Check Armstrong number (general n-digit).

```python
n = 153
temp = n
amstrng = 0

while (n!=0):
  rem = int(n%10)
  # sum = sum*10 + rem
  amstrng = amstrng + rem*rem*rem
  n = int(n/10)

if (amstrng == temp):
  print("Armstrong")

else:
  print("not amstrong")
```

6. Check Perfect number.

```python
num1 = 28
sum = 0

for i in range(1,num1):
  if (num1 % i == 0):
    # print(i)
    sum = sum+i

# print(sum)

if(sum == num1):
  print("Perfect")
else:
  print("not perfect")
```

7. Check Strong number (sum of factorial of digits).

```python
num2 = 145
temp = num2
sum = 0


while(num2!=0):
  rem = int(num2%10)
  factorial = 1
  for i in range(1,rem+1):
    factorial = factorial*i

  sum = sum + factorial
  num2 = int(num2/10);

if(sum == temp):
  print("is a strong sumber")
else:
  print("Not a strong number")
```
